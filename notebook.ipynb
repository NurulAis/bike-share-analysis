### Cleaning Data

### mengganti tipe di kolom dteday menjadi datetime

```python
#dataframe day_df
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
day_df.info()
```

```python
#dataframe hour_df
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])
hour_df.info()
```

### mengganti tipe season

```python
#mengganti kolom season di dataframe day_df
season_dict = {
    1 : 'Musim Semi',
    2 : 'Musim Panas',
    3 : 'Musim Gugur',
    4 : 'Musim Salju'
}

day_df['season'] = day_df['season'].replace(season_dict)
day_df.head()
```

```python
#mengganti kolom season di hour_df
hour_df['season'] = hour_df['season'].replace(season_dict)
hour_df.head()
```

### mengganti kolom tahun

```python
#mengganti kolom tahun di day_df
yr_dict = {
    0 : 2011,
    1 : 2012
}

day_df['yr'] = day_df['yr'].replace(yr_dict)
day_df.head()
```

```python
#mengganti kolom year di hour_df
hour_df['yr'] = hour_df['yr'].replace(yr_dict)
hour_df.head()
```

```python
workingday_dict = {
    0 : 'Working Day',
    1 : 'Not Working Day'
}

holiday_dict = {
    0 : "Holiday",
    1 : 'Not Holiday'
}

weekday_dict = {
    0 : 'Minggu',
    1 : 'Senin',
    2 : 'Selasa',
    3 : 'Rabu',
    4 : 'Kamis',
    5 : 'Jumat',
    6 : 'Sabtu'
}

weathersit_dict = {
    1 : 'Cerah',
    2 : 'Berkabut',
    3 : 'Hujan/Salju Ringan',
    4 : 'Hujan/Salju Lebat'
}

day_df['weathersit'] = day_df['weathersit'].replace(weathersit_dict)
day_df['weekday'] = day_df['weekday'].replace(weekday_dict)
day_df['holiday'] = day_df['holiday'].replace(holiday_dict)
day_df['workingday'] = day_df['workingday'].replace(workingday_dict)

hour_df['weathersit'] = hour_df['weathersit'].replace(weathersit_dict)
hour_df['weekday'] = hour_df['weekday'].replace(weekday_dict)
hour_df['holiday'] = hour_df['holiday'].replace(holiday_dict)
hour_df['workingday'] = hour_df['workingday'].replace(workingday_dict)

day_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Musim apa yang paling disukain pengguna sepeda?

```python
fav_season = day_df.groupby(by='season')['cnt'].sum().sort_values(ascending=False)
ax = sns.barplot(x=fav_season.index, y = fav_season.values)

#tambah label ditiap bar
for index, value in enumerate(fav_season.values):
    plt.text(x = index, y=value, s=f'{value}', ha='center')

plt.xlabel('Musim')
plt.ylabel('Jumlah pesepeda')
plt.title('Jumlah Pesepeda ditiap musim')
plt.show()
```

### Pertanyaan 2: Waktu yang paling disukai pengguna sepeda?

```python
bike_hour = hour_df.groupby(by='hr')['cnt'].sum()
bike_hour
x = bike_hour.index
y = bike_hour.values
plt.figure(figsize=(12,5))
plt.grid(True)
plt.plot(x,y)
plt.xticks(range(0,24))
plt.xlabel('Jam')
plt.ylabel('Jumlah Pesepeda')
plt.title('Persebaran Pesepeda Tiap Jam')
plt.show()
```

### Pertanyaan 3 : Apa hari yang paling ramai pesepeda?

```python
fav_day = day_df.groupby(by='weekday')['cnt'].sum().sort_values(ascending=False)
x = fav_day.index
y = fav_day.values

ax = sns.barplot(x=x,y=y)
for index, value in enumerate(y):
    plt.text(x = index, y = value, s=f'{value}', ha='center')

plt.xlabel('Hari')
plt.ylabel('Jumlah Pesepeda')
plt.title('Persebaran Pesepeda Berdasarkan Hari')  
plt.show()
```

### Pertanyaan 4 : Bagaimana perbandingan antara pengguna sepeda yang berlanggana dan tidak berlangganan?

```python
rc_df = day_df[['registered', 'casual']].sum()

plt.title('Distribusi Jumlah Pesepeda')
plt.pie(rc_df, 
        labels=rc_df.index,
        autopct='%1.2f%%', #menampilkan persentase dengan 2 desimal
        colors=['#ff9999', '#66b3ff'], # waarna tiap kategori
        explode=(0.1, 0), 
        shadow=True,# menambahkan bayangan
        startangle=100 #sudut awal pie chart
        )
plt.show()
```

### Pertanyaan 5 : Apa cuaca yang paling banyak pesepeda

```python
fav_weather = hour_df.groupby(by='weathersit')['cnt'].sum().sort_values(ascending=False)
x = fav_weather.index
y = fav_weather.values

ax = sns.barplot(x=x,y=y)
for index, value in enumerate(y):
    plt.text(x = index, y = value, s=f'{value}', ha='center')
plt.xticks(rotation=10) 
plt.xlabel('Cuaca')
plt.ylabel('Jumlah Pesepeda')
plt.title('Persebaran Pesepeda Berdasarkan Cuaca')  
plt.show()
```
